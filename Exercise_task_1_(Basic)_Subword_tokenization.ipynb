```python
from wordfreq import top_n_list

# Get the top 100,000 words in English ('en')
top_100k_words = top_n_list('en', 100000)

# Print first 10
print(top_100k_words[:10])


```

```python
!pip install -q wordfreq transformers

from wordfreq import top_n_list
from transformers import AutoTokenizer


```

```python
# Step 2: Load a subword tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Step 3: Find 20 most frequent words that split into multiple subwords
count = 0

for rank, word in enumerate(top_100k_words, start=1):
    tokens = tokenizer.tokenize(word)

    if len(tokens) > 1:
        print("Rank:", rank)
        print("Word:", word)
        print("Split into subwords:", tokens)
        print("-" * 40)
        count += 1

    if count == 20:
        break
```
